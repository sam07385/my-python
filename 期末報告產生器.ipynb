```python
import os, re
import random
```

```python
def 讀取JSON(fileName=""):
    import json
    if fileName!='':
        strList = fileName.split(".")
        if strList[len(strList)-1].lower() == "json":
            with open(fileName,mode='r',encoding="utf-8") as file:
                return json.loads(file.read())
```

```python
data = 讀取JSON("data.json")
名言 = data["famous"] # a 代表前話，b代表後話
前話 = data["before"] # 在名言前面廢話
後話 = data['after']  # 在名言後面廢話
廢話 = data['bosh'] # 文章主要的廢話
```

```python
xx = "超多廢話"

重複度 = 2
```

```python
def 洗牌(列表):
    global 重複度
    集合 = list(列表) * 重複度
    while True:
        random.shuffle(集合)
        for 元素 in 集合:
            yield 元素
```

```python
下一句廢話 = 洗牌(廢話)
下一句名言 = 洗牌(名言)
結尾句子 = 洗牌(後話)
```

```python
def 加上名言點綴():
    global 下一句名言
    xx = next(下一句名言)
    xx = xx.replace(  "a",random.choice(前話) )
    xx = xx.replace(  "b",random.choice(後話) )
    return xx
```

```python
def 結尾():
    global 結尾句子
    xx = next(結尾句子)
    return xx
```

```python
def 另起一段():
    xx = " "
    xx += "\r\n"
    xx += "    "
    return xx
```

```python
if __name__ == "__main__":
    xx = input("請輸入文章主題:")
    num = input("請輸入報告下限字數")
    tit = xx
    ending = "最後我想簡單做個總結。而對於這個議題整合與評估，經由本文的討論及重整後更加完備，也期許未來將有更多值得我們期待的關於此專題的想法出現！"
    for x in xx:
        tmp = str()
        while ( len(tmp) < int(num) ) :
            分支 = random.randint(0,100)
            if 分支 < 5:
                tmp += 結尾()
                tmp += 另起一段()
            elif 分支 < 20 :
                tmp += 加上名言點綴()
            else:
                tmp += next(下一句廢話)
        tmp = tmp.replace("x",xx)
        print("    "+tmp+"關於"+tit+ending)
```
